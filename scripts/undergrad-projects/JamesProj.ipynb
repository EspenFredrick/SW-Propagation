```python
def covariance(x, y):
    print(len(x))
    summationStore = []
    xBar = np.average(x) # Average value of x-array
    yBar = np.average(y)

    for i, j in zip(x, y):
        summationStore.append(((i-xBar)*(j-yBar))/len(x))

    cov = np.sum(summationStore)
    return cov
```
